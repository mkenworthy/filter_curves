```python
wasp = parse("WASP.SuperWASP.SW.xml").get_first_table()
tess = parse("TESS.TESS.Red.xml").get_first_table()
asas_g = parse("SLOAN.SDSS.g.xml").get_first_table()
asas_r = parse("SLOAN.SDSS.r.xml").get_first_table()
asas_i = parse("SLOAN.SDSS.i.xml").get_first_table()
prompt_v = ascii.read('Bessel_V-1.txt')
prompt_b = ascii.read('Bessel_B-1.txt')
prompt_r = ascii.read('Bessel_R-1.txt')
prompt_i = ascii.read('Bessel_I-1.txt')
prompt_u = ascii.read('Bessel_U-1.txt')

print(prompt_v)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,4))
plt.plot(asas_g.array['Wavelength'],asas_g.array['Transmission'],label='ASAS g')
plt.plot(asas_r.array['Wavelength'],asas_r.array['Transmission'],label='ASAS r')
plt.plot(asas_i.array['Wavelength'],asas_i.array['Transmission'],label='ASAS i')
plt.plot(prompt_v['wlen']*10., prompt_v['tx']/100,label='PROMPT V')
plt.plot(prompt_b['wlen']*10., prompt_b['tx']/100,label='PROMPT B')
plt.plot(prompt_r['wlen']*10., prompt_r['tx']/100,label='PROMPT R')
plt.plot(prompt_u['wlen']*10., prompt_u['tx']/100,label='PROMPT U')
plt.plot(prompt_i['wlen']*10., prompt_i['tx']/100,label='PROMPT I')

plt.legend()
plt.xlabel('Wavelength [Angstroms]')
plt.ylabel('transmission')
plt.title('Filters observing J1407')
plt.xlim(3500,11000)
```
